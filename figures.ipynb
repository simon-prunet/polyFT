```python
from poly import *
%matplotlib inline
from matplotlib import pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
p = petal_FT(n_border=40,n_petals=6,profile_type='arch_cos',profile_path='Matlab_files/NW2.mat')
```

```python
rEj = rot(p.Ej)
Gammap = p.Gamma - rEj*100
Gammad = p.Gamma + 2*p.Ej
```

```python
fig = plt.figure()
ax = fig.gca()
for i in range(p.npoints):
    plt.plot([p.Gamma[i,0],Gammap[i,0]],[p.Gamma[i,1],Gammap[i,1]],c='gray',alpha=0.1)
    plt.plot([p.Gamma[i,0],Gammad[i,0]],[p.Gamma[i,1],Gammad[i,1]],c='red',alpha=1)
#plt.scatter(p.Gamma[:,0],p.Gamma[:,1],s=10)
ax.axis('equal')
```

```python
fig.savefig('Boivin-demo.png')
```

### Figure comparing pixelized mask and polygonal mask

```python
p = petal_FT(n_border=100,profile_type='sister',profile_path='Matlab_files/NW2.mat')
```

```python
plt.scatter(p.Gamma[:,0],p.Gamma[:,1],s=1)
```

```python
m = p.pixelized_mask(embed_factor=2,n_pixels=4096)
```

```python
plt.viridis()
plt.scatter(p.Gamma[:,0],p.Gamma[:,1],s=0.1)
plt.plot(p.Gamma[:,0],p.Gamma[:,1],lw=.1)
plt.imshow(np.fft.fftshift(m),extent=p.pixelized_bbox())
fig=plt.gcf()

fig.savefig('pixelized-vs-polygone.pdf')
plt.clf()

plt.scatter(p.Gamma[:,0],p.Gamma[:,1],s=10)
plt.plot(p.Gamma[:,0],p.Gamma[:,1],lw=.5)
plt.imshow(np.fft.fftshift(m),extent=p.pixelized_bbox())

plt.xlim((23,25.7))
plt.ylim((23,25.7))
fig.savefig('upper-right.pdf')
```
