```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
fig,ax=plt.subplots(1,2)
arr=np.loadtxt('timing.txt',skiprows=1)
slope = arr[-1,1]/arr[-1,0]
ax[0].plot(arr[:,0],slope*arr[:,0])
ax[0].plot(arr[:,0],arr[:,1],'.',ms=10)
ax[0].set_xlabel('Number of vertices per half petal')
ax[0].set_ylabel('Compute time (s)')

arr=np.loadtxt('timing-freqs.txt')
slope = arr[-1,1]/arr[-1,0]
ax[1].plot(arr[:,0],slope*arr[:,0])
ax[1].plot(arr[:,0],arr[:,1],'.',ms=10)
ax[1].set_xlabel('Number of frequencies')
ax[1].set_ylabel('Compute time (s)')
plt.tight_layout()
```
